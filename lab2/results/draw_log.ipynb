```python
from matplotlib import pyplot as plt
import seaborn as sns
import os
```

```python
file_pth = "./m-b+dp-02+n-T+wd-005+r-T+e-240+i-192.log"
```

```python
# extract train_losses, test_losses, test_accs
train_losses = []
val_losses = []
test_accs = []
for line in open(file_pth):
    train_losses.append(float(line.split(",")[2].split(":")[1]))
    val_losses.append(float(line.split(",")[5].split(":")[1]))
    test_accs.append(float(line.split(",")[6].split(":")[1]))
```

```python
# draw
print("*"*50)
print("Draw Loss Curve")
print("*"*50)
sns.set(rc={'figure.figsize':(10, 6)})
sns.set_theme()
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5)
plt.xlabel("Epoch")
plt.ylabel("Loss")
sns.lineplot(x=range(len(train_losses)), y=train_losses, label="train")
sns.lineplot(x=range(len(val_losses)), y=val_losses, label="val")
plt.savefig("../figs/loss_curve.png")
```

```python
# draw
print("*"*50)
print("Top1 Accuracy Curve")
print("*"*50)
sns.set(rc={'figure.figsize':(10, 6)})
sns.set_theme()
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5)
plt.xlabel("Epoch")
plt.ylabel("Top1 Accuracy")
sns.lineplot(x=range(len(test_accs)), y=test_accs, label="test")
plt.savefig("../figs/acccuracy_curve.png")
```
